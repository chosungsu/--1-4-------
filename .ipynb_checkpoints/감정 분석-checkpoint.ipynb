```python
dic1 = open('dic/positive-words.ko.txt', encoding='utf-8')
dic2 = open('dic/negative-words.ko.txt', encoding='utf-8')
positive_words=[]
negative_words=[]
for line in dic1:
    positive_words.append(line.strip('\n'))
for line in dic2:
    negative_words.append(line.strip('\n'))
```

```python
import pandas as pd
df = pd.read_csv('data/forth_revolution.csv', encoding='utf-8')
df[0:10]
```

```python
title = df.get('Title')
description = df.get('Description')
result = {title[0]:description[0]}
for i in range(len(title)) :
    if result.get(title[i]) != None :
        result[title[i]] = result.get(title[i]) + "\n" + description[i]
    else :
        result[title[i]] = description[i]
result
```

```python
from nltk.corpus import gutenberg
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
```

```python
import nltk
nltk.download('punkt')
tokens = []
for f, t in result.items() :
    token = word_tokenize(t)
    tokens.append([f,token])
```

```python
sentiment_word=[]

for token in tokens:
    title=token[0]
    sentiment=0
    count=0
    
    for t in token[1]:
        if t in positive_words:
            sentiment+=1
            count+=1
        elif t in negative_words:
            sentiment -=1
            count+=1
    if count == 0 :
        sentiment_word.append([title,0.0])
    else :
        sentiment_word.append([title,sentiment/count])
    

sentiment_word[0:10]
```

```python
import numpy as np
import matplotlib.pyplot as plt

a=[]
cnt_positive = 0
cnt_negative = 0
cnt_cent = 0
for sen in sentiment_word:
    a.append(sen[1]*100)
    if sen[1]*100 == 100 :
        cnt_positive += 1
    elif sen[1]*100 == 100 * (-1):
        cnt_negative += 1
    else :
        cnt_cent += 1
X=np.arange(len(a))

plt.title("sentiment(%)",fontsize=20)
plt.bar(X,a)
plt.show()
print("긍정(%) : ", (cnt_positive / len(sentiment_word)) * 100)
print("부정(%) : ", (cnt_negative / len(sentiment_word)) * 100)
print("중립(%) : ", (cnt_cent / len(sentiment_word)) * 100)
```
